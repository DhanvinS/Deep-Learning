```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```


- **keras.datasets.mnist.load_data()**: This function loads the MNIST dataset—a classic benchmark dataset of handwritten digits (0-9) used for image classification and machine learning tasks.
- The MNIST dataset contains **70,000 grayscale images**, each 28x28 pixels: 60,000 for training and 10,000 for testing.

**Returned values:**

- The function returns a tuple of tuples:
  - **(X_train, y_train)**: Training set images (**X_train**) and their labels (**y_train**), containing 60,000 samples.
  - **(X_test, y_test)**: Test set images (**X_test**) and their labels (**y_test**), containing 10,000 samples.

**Shapes:**

- **X_train.shape:** (60000, 28, 28)
- **y_train.shape:** (60000,)
- **X_test.shape:** (10000, 28, 28)
- **y_test.shape:** (10000,)

**Summary of what you get:**

- **X_train**: Images to train your model
- **y_train**: Correct digit labels for training images
- **X_test**: Images to test your model
- **y_test**: Correct digit labels for test images

**Use case:**  
This dataset is often used to train and evaluate models for digit recognition using deep learning libraries like Keras.

Convert the 28x28 array into an one dimensional array i.e 28x28 into single array of 784

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_train_flattened
```

```python
X_test_flattened = X_test.reshape(len(X_test), 28*28)
X_test_flattened
```

```python
X_train_flattened.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
  )

model.fit(X_train_flattened, y_train, epochs = 5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[0]
```

```python
np.argmax(y_predicted[0])
```

```python
cm = tf.math.confusion_matrix(labels = y_test, predictions = np.argmax(y_predicted, axis=1))
cm
```

```python
import seaborn as sb
plt.figure(figsize = (10, 7))
sb.heatmap(cm, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
